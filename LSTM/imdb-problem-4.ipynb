```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import imdb
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.preprocessing import sequence
```

```python
max_features = 10000
max_len = 500
(xtrain, ytrain),(xtest, ytest) = imdb.load_data(num_words = max_features)

xtrain = sequence.pad_sequences(xtrain, maxlen = max_len)
xtest = sequence.pad_sequences(xtest, maxlen = max_len)

print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
model = models.Sequential()

model.add(layers.Embedding(max_features, 32))
model.add(layers.LSTM(32))
model.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
model.summary()
```

```python
history = model.fit(xtrain, ytrain, epochs = 30, batch_size = 64, validation_split = 0.2)
```

```python
model = models.Sequential()

model.add(layers.Embedding(max_features, 32))
model.add(layers.Bidirectional(layers.LSTM(32)))
model.add(layers.Dense(1, activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
model.summary()
```

```python
model.fit(xtrain, ytrain, epochs = 5,batch_size = 128, validation_split = 0.2)
```
